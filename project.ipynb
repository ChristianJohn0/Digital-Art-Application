```python
dotKV = '''\
<PaintWidget>:
    size_hint_x: 0.8 # your code here. Minimal code is required here.#
    canvas:

<MenuWidget>:
    size_hint_x: 0.2 # your code here. Most of your layout code should go here. #
    orientation: 'vertical'
    BoxLayout:
        orientation: 'horizontal'
        size_hint_y: 1
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            id: line
            on_press: if (self.state == "down"): root.setDrawingType("Straight Line")
            canvas:
                Rectangle:
                    source: "line.png"
                    pos: self.pos
                    size: self.size
            
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Dash Line")
            canvas:
                Line:
                    points: [self.x + 5, self.y + (self.height/2), self.x + self.width - 5, self.y + (self.height/2)]
                    dash_length: 10
                    dash_offset: 10
    BoxLayout:
        orientation: 'horizontal'
        size_hint_y: 1
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Free Line")
            canvas:
                Rectangle:
                    source: "freeline.png"
                    pos: [self.x, self.y]
                    size: [self.width, self.height]
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Connect Line")
            canvas:
                Rectangle:
                    source: "connect.png"
                    pos: [self.x, self.y]
                    size: [self.width, self.height]
    BoxLayout:
        orientation: 'horizontal'
        size_hint_y: 1
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Rectangle")
            canvas:
                Rectangle:
                    source: "changes.png"
                    pos: [self.x + 15, self.y + 15]
                    size: [self.width - 30, self.height - 30]
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Ellipse")
            canvas:
                Rectangle:
                    source: "eclipse.png"
                    pos: [self.x + 15, self.y]
                    size: [self.width - 30, self.height]
    BoxLayout:
        orientation: 'horizontal'
        size_hint_y: 1
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Circle")
            canvas:
                Rectangle:
                    source: "circle.png"
                    pos: [self.x + 25, self.y]
                    size: [self.width - 50, self.height]
        ToggleButton: 
            size_hint_y: 1
            allow_no_selection: False
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType("Custom")
            canvas:
                Rectangle:
                    source: "fvvbwos.png"
                    pos: [self.x + 15, self.y]
                    size: [self.width - 30, self.height]
    BoxLayout:
        orientation: 'horizontal'
        size_hint_y: 1
        ToggleButton: 
            allow_no_selection: False
            text : "Mesh"
            group: "Drawing Type"
            on_press: if (self.state == "down"): root.setDrawingType(self.text)
        Spinner:
            text: 'mode'
            values: ['points', 'line_strip', 'line_loop', 'lines', 'triangle_strip', 'triangle_fan']
            id: meshMode 
            on_text: root.change_mode(self.text)
    
    BoxLayout:
        size_hint_y: 1
        padding: "3px"
        orientation: 'vertical'
        Label:
            text:'Mesh type'
        Slider:
            id: MeshValue
            min: 0.1
            max: 10
            value_track: True
            value_track_color: [1, 0, 0, 1]
            set_system_cursor: 'hand'
            
    BoxLayout:
        size_hint_y: 1.5
        padding: "3px"
        orientation: 'vertical'
        Label:
            text:'Width'
        Slider:
            id: sliderValue
            min: 0.1
            max: 10
            value_track: True
            value_track_color: [1, 0, 0, 1]
            on_value: root.setDemoLine()
            
    BoxLayout:
        size_hint_y: 1.5
        padding: "20px"
        Label:
            id: demoLine
            canvas:
                Line:
                    points: [self.x, self.y + (self.height/2), self.x + self.width, self.y + (self.height/2)]

    BoxLayout:
        size_hint_y: 1
        orientation: 'horizontal'
        Button:
            text: 'Undo'
            on_release: root.undo()
        Button:
            text: 'Redo'
            on_release: root.redo()
        Button:
            text: 'Erase'
            on_release: root.setDrawingType("erase")
    
    Button:
        size_hint_y: 1
        text: 'Fill Canvas'
        on_release: root.fillCanvas()
    
    Label:
        size_hint_y: 1
        text: 'Screen recorder'
            
    BoxLayout:
        size_hint_y: 1
        orientation: 'horizontal'
        Button:
            text: 'Record'
            on_release: root.record()
        Button:
            text: 'Stop'
            on_release: root.stop()
        Button:
            text: 'Pause'
            on_release: root.rec.play = False

    BoxLayout:
        size_hint_y: 1
        orientation: 'horizontal'
        Button:
            text: 'Play'
            on_release: root.play_record()
        Button: 
            text: 'Play Recent'
            on_release: root.load_video()
            
    Label:
        size_hint_y: 1
        text: 'Review'
       
    BoxLayout:
        size_hint_y: 1
        orientation: 'horizontal'
        Button:
            text: 'Play'
            on_release: root.show_play_back()
        Button:
            text: 'Pause'
            on_release: root.pause()
            
    BoxLayout:
        size_hint_y: 1
        orientation: 'horizontal'
        Button:
            text: 'Save'
            on_release: root.show_save()
        Button:
            text: 'Load'
            on_release: root.show_load();
            
            
    BoxLayout:
        size_hint_y: 4
        orientation: 'vertical'
        Label:
            size_hint_y: 1
            text: 'Color Selector'
        Label:
            size_hint_y: 3
            text: ''
            id: colorSelect
            canvas:
                Color:
                    rgba: [1,1,1,1]
                Rectangle:
                    size: self.width, self.height
                    pos: self.x, self.y
                    
<ColorLoader>:
    BoxLayout:
        size: root.size
        pos: root.pos
        orientation: 'vertical'
        ColorPicker:
            id: clp
            on_color: root.color = self.color
        BoxLayout:
            size_hint_y: None
            height: 50
            Button:
                text: "Cancel"
                on_release: root.cancel()
            Button:
                text: "Select Color"
                on_release: root.load()

<SaveDialog>:
    BoxLayout:
        size: root.size
        pos: root.pos
        orientation: 'vertical'
        FileChooserIconView:
            id: filechooser
        TextInput: 
            size_hint_y: None
            height: 50
            id: saveText
            text: "untitled.png"
        BoxLayout:
            size_hint_y: None
            height: 50
            Button:
                text: "Cancel"
                on_release: root.cancel()
            Button:
                text: "Save Image"
                on_release: root.save(filechooser.path, saveText.text)

<LoadDialog>:
    BoxLayout:
        size: root.size
        pos: root.pos
        orientation: 'vertical'
        FileChooserIconView:
            id: filechooser
        BoxLayout:
            size_hint_y: None
            height: 50
            Button:
                text: "Cancel"
                on_release: root.cancel()
            Button:
                text: "Load Image"
                on_release: root.load(filechooser.path, filechooser.selection)
                
'''
from kivy.lang import Builder
Builder.load_string(dotKV)

from kivy.config import Config
Config.set('graphics', 'fullscreen', '0')
Config.set('modules', 'recorder', '')

from random import random
from kivy.app import App
from kivy.uix.textinput   import TextInput
from kivy.uix.spinner     import Spinner
from kivy.uix.boxlayout   import BoxLayout
from kivy.uix.floatlayout   import FloatLayout

from kivy.uix.widget import Widget
from kivy.uix.button import Button
from kivy.uix.bubble import Bubble
from functools import partial
from kivy.graphics import Color, Line, Rectangle, Ellipse, Mesh
from kivy.properties import ListProperty, NumericProperty, StringProperty
from kivy.uix.colorpicker import ColorPicker
from kivy.properties import ObjectProperty
from kivy.uix.popup import Popup
from kivy.clock import Clock
import math
from math import cos, sin, pi
import os
from kivy.graphics.svg import Svg
from kivy.input.recorder import Recorder

class ColorLoader(FloatLayout):
    load = ObjectProperty(None)
    cancel = ObjectProperty(None)
    color = ListProperty(None)

class SaveDialog(FloatLayout):
    save = ObjectProperty(None)
    text_input = ObjectProperty(None)
    cancel = ObjectProperty(None)

class LoadDialog(FloatLayout):
    load = ObjectProperty(None)
    cancel = ObjectProperty(None)  
    def __init__(self, **kwargs):
        FloatLayout.__init__(self, **kwargs)

###############################################################################################################################
################################################### PAINTWIDGET ###############################################################
###############################################################################################################################
class PaintWidget(Widget):  
    lineType = StringProperty("Straight Line") # current line

    def __init__ (self):
        Widget.__init__(self)
        self.lineWidth = 1              # current line width
        self.width = 0.0
        self.height = 0.0
        self.start = {'x': 0.0, 'y': 0.0}
        self.color = [1,1,1,1]
        self.erase_color = [0,0,0,1]
        self.touch_start = None
        self.touch_end = None
        self.saveInstruction = []
        self.vertex = []
           
    def on_lineType(self, *args):
        if (self.lineType != "Connect Line"): 
            self.vertex = []
    
    def on_touch_down(self, touch):
        if (touch.x < self.width):
            if (self.lineType == "Straight Line" or self.lineType == 'Custom' or self.lineType == "Dash Line"):
                self.touch_start = {'x': touch.x, 'y': touch.y} 
            elif(self.lineType == "Connect Line"):
                if len(self.vertex) == 0:
                    self.touch_start = {'x': touch.x, 'y': touch.y}
                    self.vertex.append(self.touch_start['x'])
                    self.vertex.append(self.touch_start['y'])
                else: 
                    self.touch_start = {}
                    self.touch_start['x'] = self.vertex[-2] 
                    self.touch_start['y'] = self.vertex[-1] 
                print(self.vertex)
            elif (self.lineType == "Free Line" or self.lineType == "erase"):
                with self.canvas:
                    if (self.lineType == "erase"):Color(*self.menuWidget.erase_color, mode='rgba')
                    else:Color(*self.color, mode='rgba')
                    touch.ud['line'] = Line(points=(touch.x, touch.y), width = self.lineWidth)                
            elif (self.lineType == "Rectangle"):
                self.touch_start = {'x': touch.x, 'y': touch.y}
                with self.canvas:
                    Color(*self.color, mode='rgba')
                    touch.ud['rectangle'] = Rectangle(pos=(self.start['x'], self.start['y']), size=(0,0))
            elif (self.lineType == "Ellipse" or self.lineType == 'Circle'):
                self.touch_start = {'x': touch.x, 'y': touch.y}
                with self.canvas:
                    Color(*self.color, mode='rgba')
                    touch.ud['ellipse'] = Ellipse(pos=(self.start['x'], self.start['y']), size=(0,0)) 
            elif (self.lineType == "Mesh"):
                self.touch_start = {'x': touch.x, 'y': touch.y}
                diameter = 0
                with self.canvas:
                    Color(*self.color, mode='rgba')           
                    self.mesh = self.build_mesh(self.touch_start,diameter)
            self.saveInstruction = self.canvas.children[:]
        
    def on_touch_up(self, touch):
        if (touch.x < self.width):
            if (self.lineType == "Connect Line") and self.touch_start:
                self.vertex.append(self.touch_end['x'])
                self.vertex.append(self.touch_end['y'])
        self.touch_start = None
        
    def on_touch_move(self, touch):  
        if (touch.x < self.width):
            if (self.lineType == "Straight Line") and self.touch_start:
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                with self.canvas:
                    if 'line' not in touch.ud:
                        Color(*self.color, mode='rgba')
                        touch.ud['line'] = Line(points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']], width = self.lineWidth)
                    else:
                        touch.ud['line'].points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']] 
            elif (self.lineType == "Dash Line") and self.touch_start:
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                with self.canvas:
                    if 'line' not in touch.ud:
                        Color(*self.color, mode='rgba')
                        touch.ud['line'] = Line(points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']], width = 1, dash_length = 20, dash_offset = 20)
                    else:
                        touch.ud['line'].points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']] 
            
            elif (self.lineType == "Free Line" or self.lineType == "erase"):
                touch.ud['line'].points += [touch.x, touch.y]
            elif (self.lineType == "Connect Line"):
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                with self.canvas:
                    if 'line' not in touch.ud:
                        Color(*self.color, mode='rgba')
                        touch.ud['line'] = Line(points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']], width = self.lineWidth)
                    else:
                        touch.ud['line'].points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']] 
                
            elif (self.lineType == "Rectangle"):
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                self.calcalatePoints()
                touch.ud['rectangle'].pos=(self.start['x'], self.start['y'])
                touch.ud['rectangle'].size=(self.width, self.height)

            elif (self.lineType == "Ellipse"):
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                self.calcalatePoints()
                touch.ud['ellipse'].pos=(self.start['x'], self.start['y'])
                touch.ud['ellipse'].size=(self.width, self.height)
                
            elif (self.lineType == "Circle"):
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                self.calcalatePoints()
                touch.ud['ellipse'].pos=(self.start['x'], self.start['y'])
                diameter = math.sqrt(self.width**2 + self.height**2)
                touch.ud['ellipse'].size=(diameter, diameter)
            
            elif (self.lineType == "Mesh"):
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                self.calcalatePoints()
                diameter = math.sqrt(self.width**2 + self.height**2)
                self.canvas.children.remove(self.canvas.children[-1])
                with self.canvas:
                    Color(*self.color, mode='rgba')           
                    self.mesh = self.build_mesh(self.touch_start,diameter)
            elif (self.lineType == 'Custom') and self.touch_start:
                self.touch_end = {'x': touch.x, 'y': touch.y} 
                color = (random(), random(), random(), 1)
                if self.touch_start:
                    with self.canvas:
                        Color(*color, mode='rgba')
                        touch.ud['line'] = Line(points=[self.touch_start['x'], self.touch_start['y'], self.touch_end['x'], self.touch_end['y']], width=1)
            self.saveInstruction = self.canvas.children[:]
            
    def calcalatePoints(self):
        if (self.touch_end['x'] < self.touch_start['x']):
            self.width = self.touch_start['x'] - self.touch_end['x'] 
        else: 
            self.width = self.touch_end['x'] - self.touch_start['x'] 
        if (self.touch_end['y'] < self.touch_start['y']):
            self.height = self.touch_start['y'] - self.touch_end['y'] 
        else: 
            self.height = self.touch_end['y'] - self.touch_start['y'] 
        
        self.start['x'] = min(self.touch_start['x'], self.touch_end['x'])
        self.start['y'] = min(self.touch_start['y'], self.touch_end['y'])
        
    def build_mesh(self, meshPos, meshDiameter):
        """ returns a Mesh of a rough circle. """
        vertices = []
        indices = []
        step = 10
        istep = (pi * self.menuWidget.ids.MeshValue.value) / float(step)
        for i in range(step):
            x = meshPos['x'] + cos(istep * i) * meshDiameter
            y = meshPos['y'] + sin(istep * i) * meshDiameter
            vertices.extend([x, y, 0, 0])
            indices.append(i)
        return Mesh(vertices=vertices, indices=indices)

###############################################################################################################################
################################################### MENUWIDGET ################################################################
###############################################################################################################################
class MenuWidget(BoxLayout):
    erase_color = ListProperty()
    def __init__ (self):
        BoxLayout.__init__(self)
        self.paintWidget = None
        self.count = 0
        
    def change_mode(self, mode, *largs):
        self.paintWidget.mesh.mode = mode
        
    def setDrawingType(self, in_type):
        self.paintWidget.lineType = in_type
        
##########################################################################################################################
####################################### FILL CANVAS, WIDTH DEMO LINE, COLOR DEMO #########################################
##########################################################################################################################
    def fillCanvas(self, *args):
        paint = self.paintWidget
        select = self.ids.colorSelect
        self.erase_color = paint.color[:]
        with paint.canvas:
            Color(*paint.color, mode='rgba')
            Rectangle(pos=(0, 0), size=(paint.width,paint.height))
            
    def setDemoLine(self):
        self.paintWidget.lineWidth = self.ids.sliderValue.value 
        self.ids.demoLine.canvas.clear()
        label = self.ids.demoLine
        with self.ids.demoLine.canvas:
            Color(*self.paintWidget.color, mode='rgba')
            Line(points = [label.x, label.y + (label.height/2), label.x + label.width, label.y + (label.height/2)], width = self.paintWidget.lineWidth)
    
    def setColorDemo(self):
        self.ids.colorSelect.canvas.clear()
        with self.ids.colorSelect.canvas:
            Color(*self.paintWidget.color, mode='rgba')
            Rectangle(pos = self.ids.colorSelect.pos, size = self.ids.colorSelect.size)
##########################################################################################################################
################################################### UNDO AND REDO ########################################################
##########################################################################################################################
    def undo (self, *args):
        if (len(self.paintWidget.canvas.children) > 2):
            for _ in range(3):
                self.lastInstruction = self.paintWidget.canvas.children[-1]
                self.paintWidget.canvas.children.remove(self.lastInstruction)
            self.count = self.count + 1

    def redo (self, *args):
        if (self.count > 0 and len(self.paintWidget.saveInstruction) > 2):
            temp = 3*self.count
            for index in range(3):
                addInstruction = self.paintWidget.saveInstruction[-(temp)]
                self.paintWidget.canvas.add(addInstruction)
                if (index != 2):
                    temp = temp - 1
            self.paintWidget.canvas.children.remove(self.paintWidget.canvas.children[-(2)])
            self.count = self.count - 1
            
    def show_play_back(self, *args):
        self.paintWidget.canvas.children.clear()
        self.test = len(self.paintWidget.saveInstruction)
        Clock.schedule_interval(self.play, 1/6)
    
    def record(self, *args):
        self.rec = Recorder(filename='myrecorder.kvi')
        self.rec.record = True        
        
    def stop(self, *args):
        self.rec.record = False 
    
    def recorder_loop(instance, value):
        if value is False:
            instance.play = True

    def play_record(self, *args):
        self.paintWidget.canvas.children.clear()
        self.rec.play = True
        
    def load_video(self, *args): 
        self.rec = Recorder(filename='myrecorder.kvi')
        self.rec.play = True
    
    def play(self, *args): 
        if self.test == 0:return False
        if self.pause == "paused":return False
        addInstruction = self.paintWidget.saveInstruction[-self.test]
        self.paintWidget.canvas.add(addInstruction)
        self.test = self.test - 1
        if (self.test % 3) == 0: 
            self.paintWidget.canvas.children.remove(self.paintWidget.canvas.children[-(2)])
        
    def pause(self, *args):
        self.pause = "paused"
###########################################################################################################################
######################################## POPUP STUFFS [LOAD, SAVE, COLOR PICKER] ##########################################
###########################################################################################################################
    def on_touch_up(self, touch):
        select = self.ids.colorSelect
        if ((touch.x > select.x and touch.x < (select.width + select.x)) and (touch.y > select.y and touch.y < (select.height + select.y))):
            self.show_picker()
                    
    def show_picker(self):
        self.colorPopup = ColorLoader(load=self.loadColor, cancel=self.dismiss_popup, color=self.paintWidget.color)
        self._popup = Popup(title="Pick Color", content=self.colorPopup, size_hint=(0.9, 0.9))
        self._popup.open()
        
    def dismiss_popup(self):
        self._popup.dismiss()

    def loadColor(self):
        self.paintWidget.color = self.colorPopup.color
        self.setDemoLine()
        self.setColorDemo()
        self.dismiss_popup()
        
    def save(self, path, filename):
        self.paintWidget.export_to_png(os.path.join(path, filename))
        self.dismiss_popup()
             
    def show_save(self, *args):
        content = SaveDialog(save=self.save, cancel=self.dismiss_popup)
        self._popup = Popup(title="Save file", content=content, size_hint=(0.9, 0.9))
        self._popup.open()
        
    def show_load(self, *args):
        content = LoadDialog(load=self.load, cancel=self.dismiss_popup)
        self._popup = Popup(title="Load file", content=content, size_hint=(0.9, 0.9))
        self._popup.open()

    def load(self, path, filename):
        with self.paintWidget.canvas:
            Rectangle(source=str(filename[0]), pos=self.paintWidget.pos, size=self.paintWidget.size)
        self.dismiss_popup()
        
###########################################################################################################################
###########################################################################################################################
class MyPaintApp(App):
    def build(self):
        parent = BoxLayout()
        PW = PaintWidget ()
        MW = MenuWidget ()
        LC = ColorLoader ()
        MW.paintWidget = PW
        MW.ColorLoader = LC
        PW.menuWidget = MW
        parent.add_widget (PW)
        parent.add_widget (MW)
        return parent

if __name__ == '__main__':
    MyPaintApp().run()
```
